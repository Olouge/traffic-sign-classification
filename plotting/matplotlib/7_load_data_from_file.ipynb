```python
%matplotlib inline
from pylab import *
import csv
```

```python
x = []
y = []

with open('data/7_example.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plot(x, y, label='Loaded from file!')

xlabel('x')
ylabel('y')
title('Interesting Graph\nCheck it out')

legend()
show()
```
