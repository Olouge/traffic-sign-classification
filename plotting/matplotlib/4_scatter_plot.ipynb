```python
%matplotlib inline
from pylab import *
```

```python
x = [2, 4, 6, 8, 10]
y = [22, 45, 33, 11, 44]

x2 = [1, 3, 5, 7, 9]
y2 = [43, 67, 23, 17, 73]

scatter(sorted(x+x2), sorted(y+y2), label='skitscat', color='k')
plot(sorted(x+x2), sorted(y+y2), color='r')

xlabel('x')
ylabel('y')
title('Interesting Graph\nCheck it out')

legend()
show()
```
